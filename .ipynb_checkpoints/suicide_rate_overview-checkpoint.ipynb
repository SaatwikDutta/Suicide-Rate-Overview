# which generation commited the most and least suicides??

```python
def Generation_column(x):
    suicides_per_gen=df.groupby('generation')['suicides_no'].sum().sort_values(ascending=False)
    print(f'suicide count per {suicides_per_gen}')
    
    plt.figure(figsize=(8,5))
    suicides_per_gen.plot(kind='bar')
    plt.title('generation wise suicides number')
    

Generation_column('generation')

```

<h4>Conclusion: Boomers commited the most suicides and Generation (GenZ) commited the least (as of 2016)</h4

# which year had the most and least number of suicides? also try to find out some other patterns

```python
def yearly_total_suicides(x):
    yearly_suicides=df.groupby(x)['suicides_no'].sum()

    print(f'total suicides by {yearly_suicides}')
    
    # suicides over the years
    plt.figure(figsize=(8,5))
    yearly_suicides.plot(kind='bar')
    plt.title('suicides over the years')
    plt.ylabel('total_suicides')
    plt.show()

    print("")


    # suicides over the years (sorted)
    plt.figure(figsize=(6,3))
    yearly_suicides.sort_values(ascending=False).plot(kind='bar')
    plt.title('sorted order of the above plot')
    plt.ylabel('total_suicides')
    plt.show()

    
    # global suicides over timne
    plt.figure(figsize=(10, 6))
    sns.lineplot(x='year', y='suicides_no', data=df.groupby('year')['suicides_no'].sum().reset_index())
    plt.title('Global Suicides Over Time')
    plt.ylabel('total suicides')
    plt.show()


    max_suicide_year = yearly_suicides.idxmax() #
    max_suicide_count = yearly_suicides.max()
    min_suicide_year = yearly_suicides.idxmin()
    min_suicide_count = yearly_suicides.min()
    
    print(f'So the year {max_suicide_year} had the most number of suicides of {max_suicide_count}')
    print('')
    print(f'And the year {min_suicide_year} had the least number of suicides of {min_suicide_count} (or atleast upto the time the data was collected)')

yearly_total_suicides('year')
```

<h4>Conclusion: 1999 had the most suicides and 2016 had the least. Also, the suicides peaked 1989 onwards and showing a gradual sign of descent 2010 onwards</h4>

# which sex commited the most number of suicides??

```python
def sex_wise_suicides(x):
    total_suicides_by_sex=df.groupby(x)['suicides_no'].sum()

    print(f'total number of suicides by {total_suicides_by_sex}')

    print('')
    
    plt.figure(figsize=(7,4))
    df.groupby('sex')['suicides_no'].sum().plot(kind='bar')
    plt.ylabel('Number of Suicides')
    plt.show()

sex_wise_suicides('sex')
```

<h4>Conclusion: Men commited the most number of suicides</h4>

# gdp per capita vs total suicides

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='gdp_per_capita ($)', y='suicides_no', data=df)
plt.title('GDP per capita vs total suicides')
plt.xlabel('GDP per capita')
plt.ylabel('total suicides')
```

<h4>Conclusion: Quite evident from the above plot, as GDP per capita increases, the number of total suicides tends to decrease. Countries with higher GDP per capita generally have fewer total suicides.
    
Also, one point to note: The data points are highly concentrated at the lower end of the GDP per capita axis, indicating that most countries in the dataset have lower GDP per capita</h4>
